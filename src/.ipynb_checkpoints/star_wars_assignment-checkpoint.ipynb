## Table creation.
The assignment requires to create some tables with data related to some planets and stations:
- idplanets has the identifier, the name and the coordinates
- planetsdata1 has some information related to the planet (population is one of that info). It has 92 columns with no header, one of them is the id.
- planetsdata2 has other information related to the planet. It has 74 columns with no header, one of them is the name of the planet (no id).
- planetaffinity has information about the affinity of a certain planet to the rebellion.
- evilstations has information on the positioning of empire outposts.

```python
query_id = '''
CREATE TABLE idplanets(
    id INTEGER PRIMARY KEY,
    name TEXT,
    x NUMERIC,
    y NUMERIC
)
'''
```

```python
cursor.execute(query_id)
conn.commit()
```

## Joining the data

Once we have created the tables and inserted the data in our database, we can proceed to join the tables into one to make our analysis. We are doing inner joins, so no NULL values should appear on the table.

```python
join_query='''CREATE TABLE planets AS(
    SELECT * FROM(
        SELECT * FROM idplanets a INNER JOIN planetsdata1 b USING(id)
    ) d1 INNER JOIN planetsdata2 d2 USING(name)
);
'''
```

```python
cursor.execute(join_query)
conn.commit()
```

# Checking outliers

Since no data dictionary was attached, it is very difficult to make an outlier detection, but we will check the only known variable: planet population.

```python
import pandas.io.sql as sqlio
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
query = '''SELECT * FROM planets;'''
```

```python
planets = sqlio.read_sql_query(query, conn)
```

We have 1746 planets from the database with 168 columns.

```python
planets.shape
```

We remove those planets that have more or less than three times the standard deviation from the average.

```python
planets = planets[np.abs(planets['population']-planets['population'].mean()) <= (3*planets['population'].std())]
planets = planets[~(np.abs(planets['population']-planets['population'].mean()) > (3*planets['population'].std()))]
```

We removed eight planets.

```python
planets.shape
```

As we can see on the next plots, there are plenty of outliers in the other variables, but since no data dictionary was supplied, we cannot be sure if these theoretical outliers are indeed real. Besides, checking the distributions we can see that some of the variables might need to be transformed so they could be normalized.

```python
for i in planets.columns[4:]:
    sns.boxplot(planets[i])
    plt.show()
```

# Predicting affinity

Next job is predict the affinity for a planet to join the rebellion.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor as GBR
from sklearn.metrics import mean_squared_error
from math import sqrt
```

We download the data from our database.

```python
query = '''SELECT * FROM planetaffinity;'''
```

```python
affinity = sqlio.read_sql_query(query, conn)
```

We check the distribution of the affinity from the planets. We can see that they are very low, so we are applying a log10 to "normalize" them.

```python
affinity['affinity'].plot.hist(bins=100)
plt.title('Affinity distribution')
plt.xlabel('Affinity')
plt.show()
```

```python
affinity['log_affinity']=np.log10(affinity['affinity'])
affinity['log_affinity'].plot.hist(bins=100)
plt.title('log(affinity) distribution')
plt.xlabel('log(affinity)')
plt.show()
```

We join the data to the affinities.

```python
df = planets.join(affinity, on = "id", rsuffix = "_r").drop("id_r", axis = 1)
```

```python
df = df.dropna()
```

With the 'x' and 'y' variable we can create a new variable that is the distance from the center of the galaxy.

```python
df['distance'] = np.sqrt(df['x']**2+df['y']**2)
```

```python
X = df.drop(['name','id','affinity','log_affinity'], axis = 1)
y = df['log_affinity']
```

We split the data into train and test.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 77)
```

We set the parameters for our Gradient Boosting Regressor.

```python
params = {'n_estimators': 500, 'max_depth': 5, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls', 'random_state': 77, 'max_features': 'sqrt', 'verbose': 1}
```

```python
regr = GBR(**params)
```

```python
regr.fit(X_train,y_train)
```

```python
y_pred = regr.predict(X_test)
train_pred = regr.predict(X_train)
```

```python
rmse = sqrt(mean_squared_error(y_test, y_pred))
train_rmse = sqrt(mean_squared_error(y_train, train_pred))
print("Test RMSE: %.4f" % rmse)
print("Train RMSE: %.4f" % train_rmse)
```

Test RMSE is about 0.47. This is about 1/5 of the affinity range. It is not a very good result, but the data would need more engineering.

```python
plt.figure(figsize=(12, 6))
plt.title("Absolute Error and RMSE")
plt.scatter(x = y_train, y = train_pred, label = "Train predict", marker='.')
plt.scatter(x = y_test, y = y_pred, label = "Test predict", marker='o')
plt.plot([-2.50,0],[-2.50,0],'r-', label = "Perfect predict")
plt.legend(loc="upper left")
plt.xlabel("Prediction")
plt.ylabel("Real")
plt.show()
```

Next we show the feature importances. Again, since a data dictionary was not supplied, we can comment little on these results.

```python
# Plot feature importance
feature_importance = regr.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
feature_importance = feature_importance[sorted_idx]
pos = np.arange(feature_importance.shape[0]) + .5
plt.figure(figsize=(15, 40))
plt.barh(pos, feature_importance, align='center')
plt.yticks(pos, X.columns[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Feature Importance')
plt.show()
```

# Evil outposts

Next job on the assingment is finding how many outposts are closer than 15000 GU from another outpost. To do this we will measure the euclidean distances among all the outposts and filter them according to the criteria.

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
query = '''SELECT id,x,y FROM evilstations;'''
```

```python
evil = sqlio.read_sql_query(query, conn)
```

```python
evil_coords = evil[["x","y"]]
```

```python
dist = euclidean_distances(evil_coords, evil_coords)
```

```python
close_stations = dist<15000
```

```python
station_id = evil['id'].tolist()
```

Now that we have the matrix distance and the filter applied, we only have to count and produce an output.

```python
for i in range(0,close_stations.shape[0]):
    n_stations = sum(close_stations[i,])-1
    if n_stations>0:
        print("Station", station_id[i], "has", n_stations, "stations closer than threshold.")
```

# Locating a rebel outpost

On the final assignment we have to locate an outpost to maximize the conversion of planets to the rebellion. This assignment is a bit loose on the constraints, so we are going to make some assumptions:
- Rebel outpost can only be placed on a planet.
- Communications are limited to 15.000 GUs.
- The rebel outpost has to be out of reach from an empire outpost.

We are going to start simplfying our dataset with the columns we need

```python
planets_simple = df[["name","x","y","population","affinity"]]
```

```python
planets_coords = planets_simple[["x","y"]]
```

We calculate the distance from all the planets to the empire outposts and tag those that are safe to locate our rebel station (All planets farther than 15000).

```python
dist_from_planet_to_outpost = euclidean_distances(planets_coords, evil_coords)
```

```python
outpost_near_planet = dist_from_planet_to_outpost<15000
planets_names = planets_simple['name'].tolist()
```

```python
dangerous_planet=list()
for i in range(0,dist_from_planet_to_outpost.shape[0]):
    if sum(outpost_near_planet[i,])>0:
        dangerous_planet.append(planets_names[i])
```

```python
safe_planets = [planet for planet in planets_names if planet not in dangerous_planet]
```

```python
planets_simple['isSafe'] = planets_simple['name'].isin(safe_planets)
```

```python
safe_planets = planets_simple.loc[planets_simple['isSafe']==True]
```

We measure the distance from these safe planets to all the planets and we calculate the potential of rebels on each planet (affinity * population).

```python
distance_from_safe_planets = euclidean_distances(safe_planets[["x","y"]], planets_coords)
```

```python
planets_simple["potential_rebels"] = planets_simple["population"]*planets_simple["affinity"]
```

We calculate the total population that the safe planets can convert into rebels within 15000 GUs (communication limit) and sort the results.

```python
planet_reach = list(zip(safe_planets['name'].tolist(),np.sum(planets_simple['potential_rebels'].tolist()*(distance_from_safe_planets<15000), axis=1).tolist()))
```

```python
sorted(planet_reach, key=lambda x: x[1], reverse=True)[:10]
```

The best location for a rebel outpost (with the constraints set) is **Kepler-164 c** with a total of 4183.64 millions potential rebels.
